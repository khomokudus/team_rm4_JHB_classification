```python
##removing links
import re
```

```python
def remove_links(df,col):
    df['new_message'] = df[col].apply(lambda s: re.sub(r"http\S+", "", s))
    return df
```

```python
df = remove_links(df,'message').copy()
```

```python
df.head()
```

```python
##removing elements preceeded by @ or #
def remove_word(df,col):
    df['no_@_#'] = df[col].apply(lambda s: re.sub(r"@\S+|#\S+", "", s))
    return df
```

```python
df = remove_word(df,'new_message')
```

```python
df.head()
```

```python
##removing puntuations except (')
def remove_punc(df,col):
    df['n_punc'] = df[col].apply(lambda s:''.join(i for i in s if i not in '!"#$%&\()*+,-./:;<=>?@[\\]^_`{|}~'))
    return df
```

```python
df = remove_punc(df,'no_@_#')
```

```python
df.head()
```

```python
##removing entities
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
def remove_enties(string):
    doc = nlp(string)
    no_entities = []
    entitis = [e.text for e in doc.ents]
    for t in doc:
        if t.text in entitis:
            pass
        else:
            no_entities.append(t.text)
    return ' '.join(no_entities)
```

```python
print(remove_enties('This is a text document that speaks about entities like Sweden and Nokia'))
```

```python
df['no_enties'] = df['n_punc'].apply(remove_enties)
```

```python
df.head()
```

```python
##removing numbers
def remove_numbers(df,col):
    df['no_numbers'] = df[col].apply(lambda s:''.join(i for i in s if i not in '0123456789'))
    return df
```

```python
df = remove_numbers(df,'no_enties')
```

```python
df.head()
```

```python
##removing imojis
def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
```

```python
df['no_imojis'] = df['no_numbers'].apply(deEmojify)
```

```python
df.head()
```

```python
##data splitting
from sklearn.model_selection import train_test_split
X = df['no_imojis']
y = df['sentiment']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
##importing models
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC,SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,f1_score
```

```python
##creating a pipeline for each model
txt1_clf = Pipeline([('vectorizer',TfidfVectorizer(stop_words='english',strip_accents='ascii')),('clf',LinearSVC())])
txt2_clf = Pipeline([('vectorizer',TfidfVectorizer(stop_words='english',strip_accents='ascii')),('clf',MultinomialNB())])
txt3_clf = Pipeline([('vectorizer',TfidfVectorizer(stop_words='english',strip_accents='ascii')),('clf',RandomForestClassifier(n_estimators=100))])
txt4_clf = Pipeline([('vectorizer',TfidfVectorizer(stop_words='english',strip_accents='ascii')),('clf',SVC(C=1.0,probability=True))])
```

```python
##training all the models
def fit_all(lst,X,y):
    lis = []
    for pip in lst:
        pip.fit(X,y)
        lis.append(pip)
    return lis
```

```python
trained=fit_all([txt1_clf,txt2_clf,txt3_clf,txt4_clf],X=X_train,y=y_train)
```

```python
trained
```

```python
##making predictions for each trained model

##LinearSV
pip1_prediction = trained[0].predict(X_test)
##Naive
pip2_prediction = trained[1].predict(X_test)
##Random forest
pip3_prediction = trained[2].predict(X_test)
##SVC
pip4_prediction = trained[3].predict(X_test)
```

```python
##evaluation of the models

print('LinearSV')
print(classification_report(y_test,pip1_prediction))
print('\n\n')
print('Naive')
print(classification_report(y_test,pip2_prediction))
print('\n\n')
print('Random forest')
print(classification_report(y_test,pip3_prediction))
print('\n\n')
print('SVC')
print(classification_report(y_test,pip4_prediction))
```

```python
##testing Accuracy score
print('LinearSV')
print(accuracy_score(y_test,pip1_prediction))
print('\n\n')
print('Naive')
print(accuracy_score(y_test,pip2_prediction))
print('\n\n')
print('Random forest')
print(accuracy_score(y_test,pip3_prediction))
```

```python
##LSTM implementation
max_no_words = 50000
max_sequence = 250
embedding = 100
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(num_words=max_no_words)
```

```python
tokenizer.fit_on_texts(df['no_imojis'].values)
```

```python
word_index  = tokenizer.word_index
```

```python
word_index
```

```python
len(word_index)
```

```python
X = tokenizer.texts_to_sequences(df['no_imojis'].values)
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
X = pad_sequences(X,maxlen=max_sequence)
```

```python
type(X)
```

```python
from keras.utils import to_categorical
```

```python
y = to_categorical(y,num_classes=4)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,
                                                    random_state=42)
```

```python
y.shape
```

```python
from keras import Sequential
from keras.layers import Embedding,SpatialDropout1D,LSTM,Dense
```

```python
from keras.callbacks import EarlyStopping
```

```python
model = Sequential()
```

```python
model.add(Embedding(max_no_words,embedding,input_length=X.shape[1]))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100,dropout=0.2,recurrent_dropout=0.2))
model.add(Dense(4,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=5,batch_size=50,
callbacks=[EarlyStopping(monitor='val_loss', patience=3, min_delta=0.0001)],
verbose=1,validation_split=0.1)
```

```python
y_pre=model.predict(X_test)
```

```python
y_pre=y_pre.argmax(axis=1)
```
